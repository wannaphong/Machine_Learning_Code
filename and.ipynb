```
import torch
import numpy as np
```

```
x_train=[[0,0],[0,1],[1,0],[1,1]]
y_train=[[0],[0],[0],[1]]
```

```
tensor_X = torch.stack([torch.Tensor(i) for  i in x_train])
```

```
tensor_y = torch.stack([torch.Tensor(i) for  i in y_train])
```
